```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#total numbers of schools 
total_schools = school_data_complete["school_name"].nunique(dropna=True)
#total numbers of students
total_students = school_data_complete["student_name"].count()
#total budget
total_budget = school_data["budget"].sum()
#average math score 
avg_math = school_data_complete["math_score"].mean()
#average reading score
avg_read = school_data_complete["reading_score"].mean()
#overall passing rate (overall average score)
avg_score = (avg_math + avg_read)/2
```

```python
#percentage of students with a passing math score (70 or greater)
pass_math = school_data_complete[school_data_complete["math_score"]>=70].count()["student_name"]
percentage_pass_math = (pass_math/total_students)*100

#percentage of students with a passing reading score (70 or greater)
pass_reading = school_data_complete[school_data_complete["reading_score"]>=70].count()["student_name"]
percentage_pass_reading = (pass_reading/total_students)*100
```

```python
#dataframe to hold the above results
district_summary = pd.DataFrame({"Total Schools":[total_schools], "Total Students":[total_students], "Total Budget":[total_budget], 
                                          "Average Math Score":[avg_math], "Average Reading Score":[avg_read], 
                                          "% Passing Math":[percentage_pass_math], "% Passing Reading":[percentage_pass_reading], 
                                         "% Overall Passing Rate":[avg_score]})
district_summary["Total Budget"] = district_summary["Total Budget"].astype(float).map("${:,.2f}".format)
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#set school name as index for school type 
school_type = school_data.set_index(["school_name"])["type"]
school_type
```

```python
#Total Students Per School
totalstuds_per_school = school_data_complete["school_name"].value_counts()
totalstuds_per_school
```

```python
#Total School Budget per school
total_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
total_school_budget
```

```python
#Per Student Budget
budget_per_student = total_school_budget/totalstuds_per_school
budget_per_student
```

```python
#Average Math Score
avg_math_per_school = school_data_complete.groupby(["school_name"]).mean()["math_score"]
avg_math_per_school 
```

```python
#Average Reading Score
avg_read_per_school = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
avg_read_per_school 
```

```python
# % Passing Math
passing_math = school_data_complete[(school_data_complete["math_score"]>=70)]
passing_math_per_school = passing_math.groupby(["school_name"]).count()["student_name"]
passing_math_per_school

#calculate percent 
perc_pass_math = (passing_math_per_school/totalstuds_per_school)*100
perc_pass_math
```

```python
# % Passing Reading
passing_read = school_data_complete[(school_data_complete["reading_score"]>=70)]
passing_read_per_school = passing_read.groupby(["school_name"]).count()["student_name"]
passing_read_per_school

#calculate percent 
perc_pass_read = (passing_read_per_school/totalstuds_per_school)*100
perc_pass_read
```

```python
# Overall Passing Rate (Average of the above two)
average_passing_rate = (perc_pass_math + perc_pass_read)/2
average_passing_rate
```

```python
#summary of the school data
school_summary = pd.DataFrame({"School Type":school_type,
                                 "Total Students":totalstuds_per_school,
                                 "Total School Budget":total_school_budget,
                                  "Budget per student":budget_per_student,
                                 "Average Math Score":avg_math_per_school ,
                                 "Average Reading Score":avg_read_per_school ,
                                 "% Passing Math":perc_pass_math,
                                 "% Passing Reading":perc_pass_read,
                                 "Overall Passing rate":average_passing_rate})
school_summary["Total School Budget"] = school_summary["Total School Budget"].astype(float).map("${:,.2f}".format)
school_summary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#Sort and display the top five schools in overall passing rate
top_five_schools = school_summary.sort_values(["Overall Passing rate"],ascending = False).head(5)
top_five_schools
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
#Sort and display the five worst-performing schools
worst_five_schools = school_summary.sort_values(["Overall Passing rate"],ascending = True).head(5)
worst_five_schools

```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
range_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
#add column for range to the table 
school_summary["Spending Ranges (Per Student)"] = pd.cut(school_summary["Budget per student"], range_bins, labels=group_names)
school_summary.head()
```

```python
# Create a GroupBy object based upon "View Group"
spending_summary = school_summary.groupby("Spending Ranges (Per Student)")

spending_summary["Average Math Score", "Average Reading Score", "% Passing Math", 
                "% Passing Reading", "Overall Passing rate"].mean()
```

## Scores by School Size
